상단 우측 쇼핑 연관 더보기 클릭  
크롤링을 할 때 양이 많으면 가져오기 전에 코드가 사라지는 경우가 있으므로  
class 이름이 realtedTags_btn_more__3Um4x인 요소가 로딩될 때까지 3초를 기다리고 해당 요소를 클릭하는 명령어를 실행

```python
WebDriverWait(browser, 3).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#__next > div > div.style_container__1YjHN > div > div.style_content_wrap__1PzEo > div.style_content__2T20F > div.seller_filter_area > ul > li:nth-child(2) > a")))
```

4백원 ~ 8백원 제품 필터링 클릭

```python
WebDriverWait(browser, 3).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#__next > div > div.style_container__1YjHN > div > div.filter_finder__1Gtei > div.filter_finder_filter__1DTIN > div.filter_finder_col__3ttPW.filter_finder_etc__-MpDo > div.finder_etc_inner > ul > li.filter_active__7ne8N > a")))
```
