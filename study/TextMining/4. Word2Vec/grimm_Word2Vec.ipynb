```python
import requests
import re
```

```python
req = requests.get("http://www.gutenberg.org/files/2591/2591-0.txt")
```

```python
grimm = req.text[2993:540077]
grimm
```

```python
grimm = re.sub(r'[^\w\.]', " ", grimm)
```

```python
grimm
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import sent_tokenize

sentence = sent_tokenize(grimm)
print(sentence[0])
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

data_cln = []
stop_words = set(stopwords.words('english'))

for i in sentence:
    sentence = word_tokenize(i) #단어 토큰화
    result = []
    
    for word in sentence:
        word = word.lower()
        if word not in stop_words: #불용어 제거
            if len(word) > 2:
                result.append(word)
                
    data_cln.append(result)
print(data_cln[0])
```

## Word2Vec 모델

```python
from gensim.models.word2vec import Word2Vec
```

```python
model = Word2Vec(data_cln, sg = 1, vector_size = 100, 
                window = 3, min_count = 3, workers = -1) # data_cln이라는 데이터 사용
```

```python
model.save("word2vec.model") # 만든 모델 저장
```

```python
model = Word2Vec.load("word2vec.model")
```

```python
model.wv["apple"] #출력되는 벡터의 크기 = 100
```

```python
model.wv['forest']
```
