```python
from string import punctuation
def repreprocessing(s) :
  #s = s.encode("utf8").decode("ascii","ignore")
  s = s.encode("utf8").decode("ascii","ignore") # 표준 아스키 코드(소문자)로 변환하고 모르는 것은 ignore
  return ''.join(c for c in s if c not in punctuation).lower()

text = [repreprocessing(x) for x in headline]
text[:5]
```

```python
from keras_preprocessing.text import Tokenizer

t = Tokenizer()
t.fit_on_texts(text)

vocab_size = len(t.word_index) + 1
print("Word set size : %d" % vocab_size)
print(f"Word set size2 : {vocab_size}") #print f-string
```

```python
sequences = list()
for line in text :
  encoded = t.texts_to_sequences([line])[0]
  for i in range(1, len(encoded)) :
    sequence = encoded[:i+1]
    sequences.append(sequence)
sequences[:11]
```

```python
index_to_word={}
for key, value in t.word_index.items():
  index_to_word[value] = key
print("582 : " + index_to_word[582])
print("1115 : " + index_to_word[1115])
len(index_to_word)
```

```python
max_len=max(len(I) for I in sequences)
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
sequences = pad_sequences(sequences, maxlen = max_len, padding='pre')
print(sequences[:5])
len(sequences)
# 우리가 봐야 할 입력 : 7803
```

```python
import numpy as np
sequences = np.array(sequences)
X = sequences[:,:-1] 
y = sequences[:,-1] # 문장의 가장 마지막 단어 : label(= category -> 답)
```

```python
print(X[:3])
```

```python
print(y[:3])
```

```python
# categorical : 범주화 => softmax 함수 사용
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=vocab_size)

# activation function 선택 
# sigmoid : 이진분류
# softmax : 다중분류
```

```python
from tensorflow.keras.layers import Embedding, Dense, LSTM
from tensorflow.keras.models import Sequential
#from functools import wraps
import time
start_time = time.time()
# your script
#time.sleep(5)
start = time.time()
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length=max_len-1))
model.add(LSTM(128))
model.add(Dense(vocab_size, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model.fit(X,y,epochs=200)

elapsed_time = time.time() - start_time
time.strftime("%H:%M:%S", time.gmtime(elapsed_time))

# 7083개의 문장을 200번(epoch) 학습 => weight 값을 200번 조정
# 24개의 노드가 단어 단위로 입력됨
# RNN : 선후 관계를 알려주기 위해 context node
# loss를 줄이기 위해 backpropagation 수행
```

```python
def sentence_generation(model, t, current_word, n) : 
  init_word = current_word
  sentence = ''
  for _ in range(n) :
    encoded = t.texts_to_sequences([current_word])[0]
    encoded = pad_sequences([encoded], maxlen=23, padding='pre')
    #result = model.predict_classes(encoded, verbose=0) #tf2.6 deprecated
    y_prob = model.predict(encoded, verbose=0)
    result = y_prob.argmax(axis=-1)
    for word, index in t.word_index.items() :
      if index == result:
        break
    current_word = current_word + ' ' +word
    sentence = sentence + ' ' + word
  sentence = init_word + sentence
  return sentence
```

- epoch = 200일 때  
 문장의 정확도가 높음
- epoch = 20 일 때  
 문장의 편향이 심함  
 ex) 'i'를 넣으면 'i' 근처의 문자들로 구성된 문장이 주로 출력

```python
print(sentence_generation(model, t, 'i', 10))
```

```python
print(sentence_generation(model, t, 'you', 10))
```
